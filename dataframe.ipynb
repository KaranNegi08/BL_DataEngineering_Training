```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, desc,sum
```

```python
spark = SparkSession.builder.\
    appName("saprksql").\
    getOrCreate()
```

```python
print(spark.version)
```

```python
data = spark.read.format('csv').\
        option('inferSchema','true').\
        option('header','true').\
        option('path','operation_management.csv').\
        load()
```

```python
data.printSchema()
```

```python
data2 = data.select("industry","value").\
    filter(col("value") > 4000).\
    orderBy(desc("value"))
```

```python
data2.show(5)
```

```python
data2.createOrReplaceTempView("data")
```

```python
spark.sql("select industry, value from data where value > 4000").show()
```
